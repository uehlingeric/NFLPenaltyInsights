```python
import pandas as pd
from matplotlib import pyplot as plt
import numpy as np
```

```python
drives = pd.read_csv("../data/processed/drives.csv")
drives
```

While looking at the other drive results might be interesting, I'm going to focus on the most common results.

```python
drives = drives[drives.result.isin(["Punt", "Touchdown", "Field Goal", "Interception", "Fumble", "Safety"])]
drives
```

```python
"""
Add "net_penalty_yards" column, which is a quantitative measure of how much penalties benefitted an offense or a
defense in a specific drive. Positive values mean that the offense gained net penalty yards.
"""

drives["net_penalty_yards"] = drives.total_def_pen_yards - drives.total_off_pen_yards

drives
```

```python
drives.groupby("result").net_penalty_yards.mean()
```

On the average drive that results in a touchdown, the offense **gains** an average of 2.14 yards due to penalties. On the average drive that results in a field goal, the offense **gains** an average of 0.65 yards due to penalties. In contrast, on the average drive that end in a safety, the offense **loses** an average of 2.37 yards due to penalties, and on the average drive that ends in a punt, the offense **loses** an average of 1.33 yards due to penalties. 

So, it is clear that if we look at drives that result in offensive scores versus drives that don't, it is clear that drives that have more points scored tend to have penalties that benefit the offense more than the defense.

To dig deeper, lets look at which penalties tend to have the biggest impact on drives.

```python
from sklearn.feature_selection import chi2, SelectKBest

# penalty columns
X = drives.iloc[:,12:94]

# drive results
y = drives.result

k_best_selector = SelectKBest(score_func=chi2)
k_best_selector.fit(X, y)

feature_score_pairs = zip(X.columns, k_best_selector.scores_)
sorted_feature_score_pairs = sorted(zip(X.columns, k_best_selector.scores_), key=lambda x: x[1], reverse=True)

penalties_by_importance = pd.DataFrame({"Penalties": X.columns, "Importance": k_best_selector.scores_})
penalties_by_importance = penalties_by_importance.sort_values(by='Importance', ascending=False)
penalties_by_importance.head(10)
```

Here, we take a look at which penalties tend to have the largest affect on the outcome of a drive. `sklearn.feature_selection.SelectKBest` gives features an importance score based on a univariate statistical test. In this case, we are using the chi-squared test as a scoring function.

We can see that Defensive Pass Interference has by far the most importance in determing the outcome of a drive. This makes sense because Defense Pass Interference is a spot foul, which means that the penalty can potentially cost the defense an unlimited amount of yardage.

However, we can also see that almost all of the 10 most important penalties in determining the outcome of a drive are defensive penalties! This is definitely not what I would have expected to see. The one exception is Intentional Grounding. This makes sense because intentional grounding is somewhat likely to lead to safeties, and otherwise has a major impact on an offense because it is a spot foul and a loss of down.

Lets take a look at the next 10 most important penalties:

```python
penalties_by_importance.iloc[10:20,:]
```

Offensive holding appears to be the next most impactful offensive penalty. This makes sense because offensive holding is a fairly common penalty that sets the offense back 10 yards, which can have a major impact on slowing down drives.
