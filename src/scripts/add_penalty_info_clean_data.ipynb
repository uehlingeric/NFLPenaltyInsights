### Fixed Weeks

```python
# Rohan Jaggannagari (rjaggannagari)
# One dataset had week numbers and the others had a string and we had to fix that
# Also, the NFL changed their schedule to 18 games during this time period so we had to account for that
for l, current_penalty_2 in penalties_df.iterrows():
    current_year = current_penalty_2['Year']
    if (current_penalty_2['Week'] == "Wildcard Weekend"):
        penalties_df.at[l, 'Week'] = 18
        if (current_year >= 2021):
            penalties_df.at[l, 'Week'] = 19
    elif (current_penalty_2['Week'] == "Divisional Playoffs"):
        penalties_df.at[l, 'Week'] = 19
        if (current_year >= 2021):
            penalties_df.at[l, 'Week'] = 20
    elif (current_penalty_2['Week'] == "Conference Championships"):
        penalties_df.at[l, 'Week'] = 20
        if (current_year >= 2021):
            penalties_df.at[l, 'Week'] = 21
    elif (current_penalty_2['Week'] == "Super Bowl"):
        penalties_df.at[l, 'Week'] = 21
        if (current_year >= 2021):
            penalties_df.at[l, 'Week'] = 22
```

### Final Step

```python
# Rohan Jaggannagari (rjaggannagari)
# Adding penalty informatio for each game
for g, current_penalty_2 in penalties_df.iterrows():
    year = current_penalty_2['Year']
    week = current_penalty_2['Week']
    first_team = current_penalty_2['Opp']
    second_team = current_penalty_2['Team']
    penalty = current_penalty_2['Penalty']
    
    # Need to have a foreign key to help refer to other dataset
    penalty_string = f"{year}_{week}_{first_team}_{second_team}"
    penalty_string_1 = f"{year}_{week}_{second_team}_{first_team}"
    
    index_array = game_details_df['game_id']
    index_array = numpy.array(index_array)
    
    # Increase the penalty depending on what it is
    if (numpy.isin(penalty_string, index_array)):
        index = game_details_df.index[game_details_df['game_id'] == penalty_string].tolist()[0]
        game_details_df.at[index, penalty] += 1
    elif (numpy.isin(penalty_string_1, index_array)):
        index = game_details_df.index[game_details_df['game_id'] == penalty_string_1].tolist()[0]
        game_details_df.at[index, penalty] += 1

# Write the data to a csv file
game_details_df.to_csv("p_game_details.csv", index = False)
```
